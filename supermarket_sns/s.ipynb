```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a figure and axis
plt.figure(figsize=(10, 6))

# Create a KDE plot with a rug plot
sns.kdeplot(data, bw_adjust=0.5, fill=True, color='skyblue', alpha=0.5)
sns.rugplot(data, color='brown', height=0.1)

# Customize the plot
plt.title('KDE Plot with Rug Plot')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)

# Show the plot
plt.show()
```
